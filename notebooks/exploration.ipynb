```python
import numpy as numpy
import pandas as pd 
```

```python
#data = pd.read_csv('data\raw\futurice\blogs.csv')
data = pd.read_csv('../data/raw/futurice/blogs.csv')
```

```python
def string_to_list(word: str):
    return [word]

def to_list(string_list):
    return eval(string_list)

def merge_strings(string_list):
    return ' '.join(string_list)   

# create 2 inter columns
data['title to list'] = data['title'].apply(string_to_list)

data['grid paragraph to list'] = data['grid paragraph'].apply(string_to_list)


# 
data['headings'] = data['headings'].apply(to_list)

data['content paragraph'] = data['content paragraph'].apply(to_list)

data['bullets'] = data['bullets'].apply(to_list)

data['content full'] = data['title to list'] + data['grid paragraph to list'] + data['headings'] + data['content paragraph'] + data['bullets']

data['keys'] = data['title to list'] + data['grid paragraph to list'] + data['content paragraph']

data.head()
```

```python
data.drop(columns=['title to list', 'grid paragraph to list'], inplace = True)
```

```python
data.head()
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')

def tokenize(string):
    doc = nlp(string)
    return doc
```

```python
data['headings'] = data['headings'].apply(merge_strings)
data['content paragraph'] = data['content paragraph'].apply(merge_strings)
data['bullets'] = data['bullets'].apply(merge_strings)
data['content full'] = data['content full'].apply(merge_strings)
data['keys'] = data['keys'].apply(merge_strings)
data.head()
```

```python
# tokenize 
data['content full tokens'] = data['content full'].apply(tokenize)
data['keys tokens'] = data['keys'].apply(tokenize)
data.loc[1]
```

```python
model = KeyBERT()
def keywords_extract(text):
    keywords = model.extract_keywords(text, top_n=5, keyphrase_ngram_range=(1, 1), use_mmr=True, stop_words='english', diversity=0.2)
    return [x[0] for x in keywords]

def keyphrase_extract(text):
    keyphrases = model.extract_keywords(text, top_n=4, keyphrase_ngram_range=(2, 2), use_mmr = False, stop_words='english')
    return [x[0] for x in keyphrases]
```

```python
data['keywords'] = data['keys'].apply(keywords_extract)
```

```python
data['keyphrases'] = data['keys'].apply(keyphrase_extract)
```

```python
keywords = data['keywords'].tolist()
keyphrases = data['keyphrases'].tolist()
```

```python
import pandas as pd 
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

from ast import literal_eval

df = pd.read_csv('../data/processed/futurice/ada_embedded.csv')
df_stats = pd.read_csv('../data/processed/futurice/blogs_with_analytics.csv')
df_stats = df_stats[['link', 'views']]
df_stats
```

```python
df = df.merge(df_stats, on='link', how='left')
df
```

```python
X1 = np.vstack(df['title'].apply(literal_eval).apply(np.array).values)
X2 = np.vstack(df['teaser text'].apply(literal_eval).apply(np.array).values)
```

```python
X1[0]
```

```python
X2[0]
```

```python
pca = PCA(n_components=2)
X1_pca = pca.fit_transform(X1)
X2_pca = pca.fit_transform(X2)

print(X1_pca.shape)
print(X2_pca.shape)
```

```python
X1_pca = X1_pca.tolist()
X2_pca = X2_pca.tolist()
```

```python
df['title pca'] = X1_pca
df['teaser pca'] = X2_pca
```

```python
# Extract the first and second elements of "title pca" into separate columns
df[['X1', 'Y1']] = pd.DataFrame(df['title pca'].tolist(), index=df.index)
df[['X2', 'Y2']] = pd.DataFrame(df['teaser pca'].tolist(), index=df.index)
```
