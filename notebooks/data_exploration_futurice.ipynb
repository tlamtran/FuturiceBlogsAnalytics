```python
import pandas as pd

df = pd.read_csv('../data/processed/futurice/blogs_with_analytics.csv')
```

```python
df.head(3).to_csv('analytics_demo.csv')
```

```python
# Load the list from the JSON file
import json
with open('../data/processed/futurice/keywords.json', 'r') as file:
    keywords = json.load(file)

with open('../data/processed/futurice/keyphrases.json', 'r') as file:
    keyphrase = json.load(file)

with open('../data/processed/futurice/popularity.json', 'r') as file:
    popularity = json.load(file)
```

```python
data = {
    "title": df['title'],
    'date': df['date'],
    "keywords": keywords,
    "keyphrases": keyphrase,
    "popular words": popularity
}
```
